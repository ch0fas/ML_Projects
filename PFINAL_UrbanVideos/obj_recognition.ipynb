### About

This project uses the **BDD100K** dataset for object detection. This dataset [1], created by Fisher Yu and his team at UC Berkeley, is made up of dash cam footage taken in New York City, Berkeley, downtown San Francisco and the Wider Bay Area, in the United States.

**INSTRUCTIONS TO SET UP DATA - VERY IMPORTANT**

On the root folder where this notebook is located, create an `images` folder and a `labels` folder. 

Download the images here (http://bdd-data.berkeley.edu/download.html), specifically the download labeled "Detection 2020 Labels". Extract the zip file and place the 3 folders (`train`, `val`, `test`) on the images folder.

Download the actual labels here (https://dl.cv.ethz.ch/bdd100k/data/bdd100k_det_20_labels_trainval.zip). Extract the zip and find the files named `det_train.json` and `det_val.json`. Place those two files on the labels folder.

Now you can continue

```python
# Changing names of images to fit the "name" attribute on CSV. Should only need to run once

image_folders = [r'images/train',r'images/test',r'images/val']

for i in range(3):
    for f in os.listdir(image_folders[i]):
        name, ext = os.path.splitext(f)
        if name.endswith('0000100'):
            new_name = name[:-8] + ext
            os.rename(os.path.join(image_folders[i], f), os.path.join(image_folders[i], new_name))
```

```python
# Data Loading

with open('labels/det_val.json') as f:
    train_label_data = json.load(f)

index_train = {item['name'] + '': item for item in train_label_data}
len(index_train)
```
